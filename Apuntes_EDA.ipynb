#Tipos de datos en Python

- Arreglo: Una matriz donde los datos son númericos u objetos. Su representación es (n x n)
 - La representaciión gráfica de los arreglos es similar a la de una tabla. en este caso se le da el nombre de "DataFrame".
  - Para el manejo de los DataFrame, se usa la libreria de Python llamada Pandas
   - Al importar la libreria se usa "import" para indicar que se va a importar una libreria, "numpy", indicando el nombre de la libreria, "as", para indicar un alias, y "np", el cual va a ser el alias, el cual permite que en lugar de escribir numpy, solo con indicar np, es suficiente, permitiendo usar funciones como lo es "np.array", la cual sin este alias se tendría que escribir "nympy.array"
    - Al definir un alias, se puede indicar cualquier nombre, sin embargo existen ya nombre "predefinidos", los cuales permiten identificarlos en otros códigos de forma sencilla y conocer a que libreria hacer referencia el alias usado
  - **Hacer operaciones con arreglos es usar numpy**

```python
# prompt: colocar un arreglo (5,2) donde en la primera columna coloque la primera es el valo unitario, y la segunda que es la cantidad

import numpy as np

# Crear un arreglo de (5, 2)
data = np.array([
    [10, 2],  # Valor unitario: 10, Cantidad: 2
    [5, 5],   # Valor unitario: 5, Cantidad: 5
    [20, 1],  # Valor unitario: 20, Cantidad: 1
    [15, 3],  # Valor unitario: 15, Cantidad: 3
    [8, 4]    # Valor unitario: 8, Cantidad: 4
])

data
```

```python
# prompt: Multiplicar las dos columnas del arreglo data y areglarlo para quedar en la forma (5,3)

import numpy as np
# Multiplicar las dos columnas del arreglo 'data'
multiplicacion = data[:, 0] * data[:, 1]

# Crear un nuevo arreglo con la multiplicación y las dos columnas originales
nuevo_arreglo = np.column_stack((data, multiplicacion))

# Redimensionar el arreglo a la forma (5, 3)
nuevo_arreglo = nuevo_arreglo.reshape(5, 3)

nuevo_arreglo
```

```python
# prompt: Carlcular el promedio de la columna 0

import numpy as np
promedio_columna_0 = np.mean(data[:, 0])
print("El promedio de la columna 0 es:", promedio_columna_0)
```

```python
promedio_columna1=np.mean(data[:,1])
print(promedio_columna1)
promedio_columna1=np.min(data[:,1])
print(promedio_columna1)
promedio_columna1=np.max(data[:,1])
print(promedio_columna1)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Histograma de edades
sns.histplot(df['Edad'], bins=5, kde=True, color='aqua')
plt.title("Distribución de Edad")
plt.show()

# Gráfico de dispersión: Nota_Matemáticas vs Nota_Ciencias
sns.scatterplot(x="Nota_Matemáticas", y="Nota_Ciencias", hue="Género", data=df)
plt.title("Relación entre Matemáticas y Ciencias")
plt.show()

# Boxplot de notas por género
sns.boxplot(x="Género", y="Nota_Matemáticas", data=df)
plt.title("Distribución de Notas de Matemáticas por Género")
plt.show()
```
